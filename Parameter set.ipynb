```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Set the aminoacid parameter for the sequence file.

```python
R6df = pd.read_csv('~/OneDrive/python3/aminomap/Amyloid_prediction/amyloid_sequence_01.csv')
R6df.tail()
```

```python
site=R6df["Sequence"]
site.head()
```

```python
First=site.str[0]
Second=site.str[1]
Third=site.str[2]
Fourth=site.str[3]
Fifth=site.str[4]
Sixth=site.str[5]
```

```python
df1=pd.DataFrame(First)
df1=df1.rename(columns={"Sequence":"First"})
df2=pd.DataFrame(Second)
df2=df2.rename(columns={"Sequence":"Second"})
df3=pd.DataFrame(Third)
df3=df3.rename(columns={"Sequence":"Third"})
df4=pd.DataFrame(Fourth)
df4=df4.rename(columns={"Sequence":"Fourth"})
df5=pd.DataFrame(Fifth)
df5=df5.rename(columns={"Sequence":"Fifth"})
df6=pd.DataFrame(Sixth)
df6=df6.rename(columns={"Sequence":"Sixth"})
```

```python
df=pd.concat([df1,df2,df3,df4,df5,df6],axis=1)
df.head()
```

```python
R6df = pd.concat([R6df,df],axis=1)
```

```python
R6df.head()
```

```python
R6df=R6df.reset_index()
R6df
```

```python
AA = pd.read_csv('~/OneDrive/python3/aminomap/Amyloid_prediction/aminoacid_parameter_minmax.csv', index_col=0)
AA
```

```python
AA=AA.rename(columns={"0":"s"})
AA=AA.rename(columns={"1":"p"})
AA=AA.rename(columns={"2":"v"})
AA=AA.rename(columns={"3":"h"})
AA=AA.rename(columns={"4":"i"})
AA=AA.rename(columns={"5":"a"})
AA=AA.rename(columns={"6":"b"})
AA
```

```python
R6df1=pd.merge(R6df,AA,left_on="First",right_on="Name")
R6df1=R6df1.sort_values(by=["index"])
R6df2=pd.merge(R6df,AA,left_on="Second",right_on="Name")
R6df2=R6df2.sort_values(by=["index"])
R6df3=pd.merge(R6df,AA,left_on="Third",right_on="Name")
R6df3=R6df3.sort_values(by=["index"])
R6df4=pd.merge(R6df,AA,left_on="Fourth",right_on="Name")
R6df4=R6df4.sort_values(by=["index"])
R6df5=pd.merge(R6df,AA,left_on="Fifth",right_on="Name")
R6df5=R6df5.sort_values(by=["index"])
R6df6=pd.merge(R6df,AA,left_on="Sixth",right_on="Name")
R6df6=R6df6.sort_values(by=["index"])
```
